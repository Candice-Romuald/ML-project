```r
train <- read.csv(file = '../../Data/training_data.csv')
train <- train[,-3]
test <-  read.csv(file = '../../Data/test_data.csv')
test <- test [,-1]

Y <- train[,2]
train <- train[,-2]
head(train)
head(test)
```

```r
col.name <- colnames(train)
for (i in 1:20){
    plot(train[,i],train[,i+1],xlab = col.name[i],ylab = col.name[i+1])
    points(test[,i],test[,i+1],col = 2)
}
```

We see that in the test sample there is only high Intensity

```r
plot((train[,'Intensity']),Y)
```

We see that it does not really matter as the two class have about the same distribution for the response 

-> we could thus almost remove Intensity from our data.
Estonishing that it is the most important variable in boosting...

### Let s try to plot some variables

we will look to important variable in RF or Boost and try to pre process (feature) the data to improve performance by comparing the training with test data. My idea is to remove aberrant values

We will get the most important variables in a saved file computed in the boost file (see src/Non linear methods/boosting).

```r
imp <- read.csv(file = '../../Data/ImportanceBoost.csv')
dim(imp)
imp <- imp[1:1000,]
```

```r
#plot most important variable and try to identify if we can remove aberrant values

for (i in imp[1:10,'Feature']) {
    plot(train[,i],Y,xlab = i,ylab = 'V.P')
    points(test[,i],rep(1,68),col = 2)
}
```

```r
var <- imp[2,'Feature']
plot(train[,as.character(var)],Y,xlab = var,ylab = 'V.P')
points(test[,as.character(var)],rep(1,68),col = 2)
```

```r
var <- imp[3,'Feature']
plot(train[,var],Y,xlab = var,ylab = 'V.P')
points(test[,var],rep(1,68),col = 2)
```
