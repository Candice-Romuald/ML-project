```r
train <- read.csv(file = '../../Data/training_data.csv')
train <- train[,-3]
test <-  read.csv(file = '../../Data/test_data.csv')
test <- test [,-1]

Y <- train[,2]
train <- train[,-2]
head(train)
head(test)
```

```r
col.name <- colnames(train)
for (i in 1:20){
    plot(train[,i],train[,i+1],xlab = col.name[i],ylab = col.name[i+1])
    points(test[,i],test[,i+1],col = 2)
}
```

We see that in the test sample there is only high Intensity

```r
plot((train[,'Intensity']),Y)
```

We see that it does not really matter as the two class have about the same distribution for the response 

-> we could thus almost remove Intensity from our data
